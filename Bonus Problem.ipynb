```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
X = df.iloc[:,[2,4]].values
wcss = []
for i in range(1, 11):
  Km = KMeans(n_clusters = i, init='k-means++', random_state=0)
  Km.fit(X)
  wcss.append(Km.inertia_)
plt.plot(range(1, 11), wcss)
plt.title("Elbow Graph")
plt.xlabel('K(Number of clusters)')
plt.ylabel('WCSS')
```

```python
model = KMeans(n_clusters = 4, init='k-means++', random_state=0)
y = model.fit_predict(X)

plt.scatter(X[y == 0, 0], X[y == 0, 1], s=80, c = 'red', label='Cluster 1')
plt.scatter(X[y == 1, 0], X[y == 1, 1], s=80, c = 'blue', label='Cluster 2')
plt.scatter(X[y == 2, 0], X[y == 2, 1], s=80, c = 'yellow', label='Cluster 3')
plt.scatter(X[y == 3, 0], X[y == 3, 1], s=80, c = 'cyan', label='Cluster 4')


plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.legend()
```

What type of machine learning problem is this?
unsupervised clustering

segmentation tasks? K-Means Clustering which categorizes data into groups.
