```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv("Weather Dataset.csv")
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
mask = df['Temperature (C)'] > -50
df = df.sort_values(by=['Formatted Date'])
df = df[mask]

x = df['Formatted Date']
y = df['Temperature (C)']
plt.xlabel('Time')
plt.ylabel('Temperature (C)')

plt.plot(x, y)
plt.show()




```

```python
h = df['Temperature (C)']
plt.xlabel('Temperature (C)')
plt.hist(h)
plt.show()
```

```python
hmask = df['Humidity'] > 0
df = df[hmask]
a = df['Humidity']
b = df['Temperature (C)']
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
plt.scatter(a, b, s=5)
plt.show()
```

```python
import seaborn as sb

corr = df.drop(['Formatted Date', 'Summary', 'Precip Type', 'Daily Summary', 'Loud Cover'], axis=1).corr()
sb.heatmap(corr,  cmap="YlGnBu", annot=True)
plt.show()
```
